```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sys
```

```python
savepath = '/media/basic/ssd256/VinBigData_CXR/vinbigdata/train.csv'
df = pd.read_csv(savepath)
```

```python
df.head()
```

```python
df.shape
```

```python
# create folds by patid
trndf = df.copy()
folddf = trndf['image_id'].reset_index(drop=True).drop_duplicates().reset_index()
folddf['fold'] = (folddf['index'].values) % 4
folddf = folddf.drop('index', 1)
trndf = trndf.merge(folddf, on='image_id', how='inner')
```

```python
trndf.head()
```

```python
trndf.shape
```

```python
trndf[trndf.image_id == trndf.image_id.unique()[11]]
```

```python
trndf.fold.value_counts()
```

```python
# savepath = '/media/basic/ssd256/VinBigData_CXR/vinbigdata/train_folds.csv'
# trndf.to_csv(savepath, index=False)
```

```python
# check the distribution of each class in each fold
nfolds = [0, 1, 2, 3] 
for cat in trndf['class_id'].unique():
	print(cat)
	cat_samples = trndf[trndf['class_id'] == cat].reset_index(drop=True)
	for fold in nfolds:
		print(fold, len(cat_samples[cat_samples.fold == fold]))
```
